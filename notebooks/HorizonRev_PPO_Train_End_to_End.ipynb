# HorizonRev MLP Training (End-to-End)

This notebook runs NumPy MLP policy training from a fresh clone of the repository.

It will:
1. Clone the repo
2. Install lightweight training dependencies
3. Run `scripts/train_mlp.py`
4. Verify output artifacts

```python
# Set your repository URL.
REPO_URL = "https://github.com/sivagirish81/HorizonRev.git"
REPO_DIR = "horizonrev_repo"

!rm -rf $REPO_DIR
!git clone $REPO_URL $REPO_DIR
%cd $REPO_DIR
```

```python
# Install lightweight dependencies for NumPy MLP training.
!python -m pip install --upgrade pip
!python -m pip install -r requirements-train.txt
!python -m pip install -e .
```

```python
# Training configuration.
# For a quick smoke test, use train_episodes=20 and eval_seed_count=20.
train_episodes = 1200
eval_seed_count = 300
reward_mode = "uncapped"
report_style = "structured"
```

```python
# Run MLP training + evaluation.
!python scripts/train_mlp.py \
  --train-episodes $train_episodes \
  --eval-seed-count $eval_seed_count \
  --reward-mode "$reward_mode" \
  --report-style "$report_style" \
  --policy-out artifacts/trained_policy_mlp.npz \
  --policy-metadata-out artifacts/policy_metadata.json \
  --metrics-out artifacts/evaluation_metrics.json
```

```python
# Verify output artifacts.
!ls -lh artifacts

import json
from pathlib import Path

metrics_path = Path("artifacts/evaluation_metrics.json")
policy_path = Path("artifacts/trained_policy_mlp.npz")
meta_path = Path("artifacts/policy_metadata.json")

print("policy exists:", policy_path.exists())
print("metadata exists:", meta_path.exists())
print("metrics exists:", metrics_path.exists())

if meta_path.exists():
    print("\nPolicy metadata:")
    print(meta_path.read_text())

if metrics_path.exists():
    metrics = json.loads(metrics_path.read_text())
    print("\nTraining reward stats:")
    print(json.dumps(metrics.get("training_reward_stats", {}), indent=2))
```
